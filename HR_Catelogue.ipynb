### Percentage of Employee in each group

```python
# Define the age group ranges
age_groups = pd.cut(df['Age'], bins=[18, 25, 35, 45, 60, float('inf')], labels=['18-25', '26-35', '36-45', '46-55', '56+'])

# Assign the age group to a new column
df['age_group'] = age_groups

# Calculate the percentage of employees in each age group
age_group_counts = df['age_group'].value_counts(normalize=True) * 100

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
# Plotting the percentage of employees in each age group
ax = age_group_counts.plot(kind='bar')
plt.xlabel('Age Group')
plt.ylabel('Percentage')
plt.title('Percentage of Employees in Each Age Group')
for i, v in enumerate(age_group_counts):
    ax.text(i, v + 1, f"{v:.1f}%", ha='center')

plt.show()
```

### Attrition Issue

```python
# Group the DataFrame by 'educationfield' and calculate the percentage of 'Yes' and 'No' values
grouped = df.groupby('EducationField')['Attrition'].value_counts(normalize=True) * 100

# Reset the index of the grouped DataFrame to access the columns
grouped = grouped.reset_index(name='percentage')

# Filter the 'No' and 'Yes' values separately
yes_values = grouped[grouped['Attrition'] == 'Yes']
no_values = grouped[grouped['Attrition'] == 'No']


# Plotting the percentages
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed
plt.bar(no_values['EducationField'], no_values['percentage'], label='No')
plt.bar(yes_values['EducationField'], yes_values['percentage'], label='Yes')
plt.xlabel('Education Field')
plt.ylabel('Percentage')
plt.title('Attrition Percentage by Education Field')
plt.legend()
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()  # Adjust the spacing
plt.show()
```

Is is clear from the figure, that most people from Human Resources department and Technical Degree has left or willing to left the company.
